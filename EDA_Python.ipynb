```python
data = data.drop(['S.No.'], axis = 1)
data.info()
```

```python
from datetime import date
date.today().year
data['Car_Age']=date.today().year-data['Year']
data.head()
```

```python
data['Brand'] = data.Name.str.split().str.get(0)
data['Model'] = data.Name.str.split().str.get(1) + data.Name.str.split().str.get(2)
data[['Name','Brand','Model']]
```

```python
print(data.Brand.unique())
print(data.Brand.nunique())
```

```python
data['Mileage_Value'] = data['Mileage'].str.extract(r'([0-9.]+)').astype(float)
data['Mileage_Unit'] = data['Mileage'].str.extract(r'([a-zA-Z/]+)')
data['Trim'] = data['Name'].str.split().str[-1].str.lower()
def categorize_trim(trim):
    if any(x in trim for x in ['base', 'std', 'lxi', 'era', 'gle', 'dlx']):
        return 'Base'
    elif any(x in trim for x in ['sport', 'sports', 'sportz', 'active', 'sx', 'vxi', 'trend']):
        return 'Sport'
    elif any(x in trim for x in ['top', 'zxi', 'zdi', 'limited', 'platinum', 'exclusive', 'plus']):
        return 'Top'
    else:
        return 'Other'
data['Trim_Category'] = data['Trim'].apply(categorize_trim)
data[['Name', 'Mileage', 'Mileage_Value', 'Mileage_Unit', 'Trim', 'Trim_Category']].head(10)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("used_cars_data.csv")
categories = ['Fuel_Type', 'Owner_Type']
for col in categories:
    values = data[col].value_counts()
    values.plot(kind='bar', color='skyblue')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("used_cars_data.csv")
categories = ['Fuel_Type', 'Owner_Type']
for col in categories:
    values = data[col].value_counts()
    plt.figure(figsize=(7, 7))
    values.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
    plt.title(f"Distribution of {col}")
    plt.ylabel('')
    plt.tight_layout()
    plt.show()
```
